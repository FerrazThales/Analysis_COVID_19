# Panorama da COVID-19

COVID-19 √© uma doen√ßa infecciosa causada pelo v√≠rus **SARS-CoV-2**. As primeiras amostras desta doen√ßa foram encontradas na cidade de **Wuhan** (China) em Janeiro de **2020**. Dado o alto grau de transmissibilidade e letalidade deste v√≠rus, as principais na√ß√µes do mundo decretaram **medidas restritivas** severas e a obrigatoriedade do uso de **m√°scaras** para tentar conter o avan√ßo desta doen√ßa.

Transmitido principalmente por meio de **got√≠culas** provenientes de tosses ou espirros de *pessoas infectadas*, os principais sintomas desta enfermidade s√£o:

* Febre ou calafrios
* Tosse 
* Falta de Ar
* Dores de garganta e no Corpo
* Fadiga
* Anosmia (aus√™ncia de olfato)
* Mais sintomas podem ser vistos em: [Sintomas Covid]('https://www.tuasaude.com/coronavirus/)

A **detec√ß√£o** deste v√≠rus, ocorre principalmente atrav√©s de exames **PCR**. Estes exames medem a quantidade de uma prote√≠na (C-reativa) no corpo dos seres humanos. Quando est√° havendo um processo inflamat√≥rio ou **infeccioso** no corpo, h√° um aumento consider√°vel desta prote√≠na. Com isso, s√£o coletados materiais infecciosos provenientes da garganta ou nariz do paciente. Para an√°lise destas amostras se utiliza biologia molecular para encontrar os **materiais gen√©ticos** do v√≠rus da COVID-19.

### Dicion√°rio de Vari√°veis
* *iso code* : c√≥digos de tr√™s-letras da Organiza√ß√£o Internacional para Padroniza√ß√£o (ISO) para representar localidades [C√≥digos ISO - Como funcionam na pr√°tica?](https://pt.wikipedia.org/wiki/ISO_3166-1_alfa-3)
* *continent* : Continente em que est√° localizado o Pa√≠s
* *location* : Pa√≠s
* *total_cases* : total de casos confirmados com COVID-19
* *new_cases* : novos casos de COVID-19 confirmados
* *new_cases_smoothed* : m√©dia m√≥vel (7 dias) de novos casos
* *total_deaths* : total de mortes confirmadas relacionadas a COVID-19
* *new_deaths* : novos mortes relacionadas COVID-19 confirmadas
* *new_deaths_smoothed* : m√©dia m√≥vel (7 dias) de novas mortes relacionadas a COVID-19
* *total_cases_per_million* : total de casos de COVID-19 por milh√£o de pessoas
* *new_cases_per_million* : total de novos casos de COVID-19 por milh√£o de pessoas
* *new_cases_smoothed_per_million* : m√©dia m√≥vel (7 dias) de novos casos por milh√£o de pessoas
* *total_deaths_per_million* : total de mortes relacionadas ao COVID-19 por milh√£o de pessoas
* *new_deaths_per_million* : total de novas mortes relacionadas ao COVID-19 por milh√£o de pessoas
* *new_deaths_smoothed_per_million* : m√©dial m√≥vel (7 dias) do n√∫mero de mortes relacionadas a COVID-19 por milh√£o de pessoas
* *reproduction_rate* : estimativa da taxa de reprodu√ß√£o do v√≠rus da COVID-19, ou seja, a quantidade de pessoas que um infectado pode contagiar
* *icu_patients* : n√∫mero de pacientes com COVID-19 na UTI no dia registrado
* *icu_patients_per_million* : n√∫mero de pacientes com COVID-19 na UTI no dia registrado por milh√£o de pessoas
* *hosp_patients* : n√∫mero de pacientes com COVID-19 no dia registrado
* *hosp_patients_per_million* : n√∫mero de pacientes com COVID-19 no dia registrado por milh√£o de pessoas
* *weekly_icu_admissions* : n√∫mero de novos pacientes com COVID-19 que ingressaram na UTI na semana registrada
* *weekly_icu_admissions_per_million* : n√∫mero de novos pacientes com COVID-19 que ingressaram na UTI na semana registrada por milh√£o de pessoas
* *weekly_hosp_admissions* : n√∫mero de novos pacientes com COVID-19 que ingressaram em hospitais na semana registrada
* *weekly_hosp_admissions_per_million* : n√∫mero de novos pacientes com COVID-19 que ingressaram em hospitais na semana registrada por milh√£o de pessoas
* *new_tests* : novos testes de COVID-19 realizados(apenas calculados em dias consecutivos)
* *total_tests* : total de testes realizado de COVID-19
* *total_tests_per_thousand* : total de testes realizados de COVID-19 por mil pessoas
* *new_tests_per_thousand* : total de novos testes realizados de COVID-19 por mil pessoas
* *new_tests_smoothed* : m√©dia m√≥vel (7 dias) de novos testes registrados
* *new_tests_smoothed_per_thousand* : m√©dia m√≥vel (7 dias) de novos testes registrados por milhares de pessoas
* *positive_rate* : taxa de COVID-19 que deram resultados positivos para a doen√ßa
* *tests_per_case* : quantidade de testes de COVID-19 realizados para cada caso positivo
* *tests_units* : unidade de medida usada pelo pa√≠s para se referenciar aos testes
* *total_vaccinations* : total de doses de vacinas administradas
* *people_vaccinated* : quantidade de pessoas que receberam ao menos uma dose da vacina
* *people_fully_vaccinated* : quantidade de pessoas que receberam todas as doses que foram prescritas nos protocolas das vacinas
* *total_boosters* : n√∫mero de doses de refor√ßo da vacina, ou seja, doses que v√£o al√©m das prescritas pelos protocolos
* *new_vaccinations* : novas doses de vacinas aplicadas (somente considerando os dias consecutivos)
* *new_vaccinations_smoothed* : m√©dia m√≥vel (7 dias) da aplica√ß√£o de novas doses de vacina
* *total_vaccinations_per_hundred* : total de doses de vacinas administradas a cada 100 pessoas
* *people_vaccinated_per_hundred* : quantidade de pessoas que receberam ao menos uma dose da vacina a cade 100 pessoas
* *people_fully_vaccinated_per_hundred* : quantidades de pessoas que receberam todas as doses de acordo com o protocolo das vacinas a cada 100 pessoas
* *total_boosters_per_hundred* : total de doses de refor√ßo (doses que v√£o al√©m das protocolares) a cada 100 pessoas
* *new_vaccinations_smoothed_per_million* : m√©dia m√≥vel (7 dias) de novas doses administradas a cada 1 milh√£o de pessoas
* *stringency_index* : √çndice que avalia a capacidade de resposta do governo para tentar conter a pandemia (ex: lockdown, fechamento de bares, etc). Este √≠ndice vai de 0 a 100
* *population* : Popula√ß√£o total medida em 2020
* *population_density* : Densidade demogr√°fica
* *median_age* : M√©dia de idade da popula√ß√£o baseado na proje√ß√£o das Na√ß√µes Unidas para o ano de 2020
* *aged_65_older* : fra√ß√£o da popula√ß√£o que tem 65 anos ou mais
* *aged_70_older* : fra√ß√£o da popula√ß√£o que tem 70 anos ou mais
* *gdp_per_capita* : Renda per capita
* *extreme_poverty* : fra√ß√£o da popula√ß√£o que vive em condi√ß√µes de extrema pobreaza
* *cardiovasc_death_rate* : taxa de mortes ocasionadas por doen√ßas cardiovasculares a cade 100 mil pessoas
* *diabetes_prevalence* : Porcentagem da popula√ß√£o (entre 20 e 79 anos) diagnosticadas com diabetes em 2017
* *female_smokers* : fra√ß√£o de mulheres que fumam
* *male_smokers* : fra√ß√£o de homens que fumam
* *handwashing_facilities* : fra√ß√£o da popula√ß√£o que tem acesso a condi√ß√µes m√≠nimas de higiene
* *hospital_beds_per_thousand* : leitos de hospital dispon√≠vel a cada mil pessoas
* *life_expectancy* : Expectativa de vida
* *human_development_index* : √çndice de Desenvolvimento Humano (IDH)
* *excess_mortality* : excesso de mortalidade, ou seja, √© a diferen√ßa entre o n√∫mero de mortes observados em um tempo espec√≠fico e o n√∫mero de mortes que eram esperadas no mesmo per√≠odo. [Defini√ß√£o de excess mortality](https://www.cdc.gov/nchs/nvss/vsrr/covid19/excess_deaths.html)
* *date* : data de registro da informa√ß√£o

*Como o dataset abrange muitas datas, inicialmente vamos analisar um dia espec√≠fico **(31/08/21)** para verificar a rela√ß√£o entre o n√∫mero de **infectados** (total de casos) e n√∫mero de mortes relacionadas ao COVID-19.*

*Este dia tem boa representatividade, j√° que est√° h√° aproximadamente 02 anos ap√≥s a **primeira** detec√ß√£o da doen√ßa na China.*

```
#definindo uma fun√ß√£o para colocar os r√≥tulos de dados em cima de cada barra
def addlabels(x,y,eixo):
  for i in range(len(x)):
    if eixo == 0:
      #se for do primeiro gr√°fico temos de adicionar um valor maior para o Y, porque existem mais casos do que mortes
      ax[eixo].text(i,y[i] + 1000000,'{:,}'.format(int(y[i])).replace(',','.'),ha='center',fontsize=14)
    else:
      #se for do segundo gr√°fico temos de adicionar um valor menor para o Y, porque existem menos mortes do que casos de COVID-19
      ax[eixo].text(i,y[i] + 20000,'{:,}'.format(int(y[i])).replace(',','.'),ha='center',fontsize=14)
```

```
#Avaliando os 10 paises tem mais casos no dia 31/08/2021
dia = '31/08/2021'

#criando o figure e o ax no matplotlib
fig,ax = plt.subplots(figsize=(20,8),nrows=2,ncols=1)

#preparando os dados dos pa√≠ses com mais casos
df_paises_com_mais_casos = df_lista_sem_continentes.loc['2021-08-31'][['location','total_cases']].sort_values(
    by='total_cases',ascending=False)[:10]

#preparando os dados dos pa√≠ses com mais mortes
df_paises_com_mais_mortes = df_lista_sem_continentes.loc['2021-08-31'][['location','total_deaths']].sort_values(
    by='total_deaths',ascending=False)[:10]


#ajustando os par√¢metros para o primeiro gr√°fico
ax[0].bar(data=df_paises_com_mais_casos,x='location',height='total_cases',color='grey')
ax[0].set_title('Casos acumulados de COVID-19 at√© ' + dia,loc='left',pad=25,fontdict={'fontsize':18})
ax[0].bar(data=df_paises_com_mais_casos[df_paises_com_mais_casos['location'] == 'Brazil'],x='location',height='total_cases',color=(0.2, 0.4, 0.6, 0.6))
ax[0].tick_params(axis='x',labelsize=13,pad=10)

#colocando o r√≥tulo de dados para o primeiro gr√°fico
addlabels(df_paises_com_mais_casos['location'],df_paises_com_mais_casos['total_cases'],0)

#retirando o eixo y e background de fundo
ax[0].set_frame_on(False)
ax[0].get_yaxis().set_visible(False)

#ajustando os par√¢metros para o segundo gr√°fico
ax[1].bar(data=df_paises_com_mais_mortes,x='location',height='total_deaths',color='grey')
ax[1].set_title('Casos de mortes relacionadas a COVID-19 acumuladas at√© '+ dia,loc='left',pad=25,fontdict={'fontsize':18,'Family':'Dejavu Sans'})
ax[1].bar(data=df_paises_com_mais_mortes[df_paises_com_mais_mortes['location'] == 'Brazil'],x='location',height='total_deaths',color=(0.2, 0.4, 0.6, 0.6))
ax[1].tick_params(axis='x',labelsize=13,pad=10)

#colocando o r√≥tulo de dados para o segundo gr√°fico
addlabels(df_paises_com_mais_mortes['location'],df_paises_com_mais_mortes['total_deaths'],1)

#retirando o eixo y e background de fundo
ax[1].get_yaxis().set_visible(False)
ax[1].set_frame_on(False)

#plotando os gr√°ficos
plt.tight_layout(pad=4.0);
```
