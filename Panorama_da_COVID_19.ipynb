# Panorama da COVID-19

COVID-19 é uma doença infecciosa causada pelo vírus **SARS-CoV-2**. As primeiras amostras desta doença foram encontradas na cidade de **Wuhan** (China) em Dezembro de **2019**. Dado o alto grau de transmissibilidade e letalidade deste vírus, as principais nações do mundo decretaram **medidas restritivas** severas e a obrigatoriedade do uso de **máscaras** para tentar conter o avanço desta doença.

Transmitido principalmente por meio de **gotículas** provenientes de tosses ou espirros de *pessoas infectadas*, os principais sintomas desta enfermidade são:

* Febre ou calafrios
* Tosse 
* Falta de Ar
* Dores de garganta e no Corpo
* Fadiga
* Anosmia (ausência de olfato)
* Mais sintomas podem ser vistos em: [Sintomas Covid]('https://www.tuasaude.com/coronavirus/)

A detecção deste vírus, ocorre principalmente através de exames **PCR**. Estes exames medem a quantidade de uma proteína (C-reativa) no corpo dos seres humanos. Quando está havendo um processo inflamatório ou infeccioso no corpo, há um aumento considerável desta proteína. Com isso, são coletados materiais infecciosos provenientes da garganta ou nariz do paciente. Para análise destas amostras se utiliza biologia molecular para encontrar os materiais genéticos do vírus da COVID-19.

### Dicionário de Variáveis
* *iso code* : códigos de três-letras da Organização Internacional para Padronização (ISO) para representar localidades [Códigos ISO - Como funcionam na prática?](https://pt.wikipedia.org/wiki/ISO_3166-1_alfa-3)
* *continent* : Continente em que está localizado o País
* *location* : País
* *total_cases* : total de casos confirmados com COVID-19
* *new_cases* : novos casos de COVID-19 confirmados
* *new_cases_smoothed* : média móvel (7 dias) de novos casos
* *total_deaths* : total de mortes confirmadas relacionadas a COVID-19
* *new_deaths* : novos mortes relacionadas COVID-19 confirmadas
* *new_deaths_smoothed* : média móvel (7 dias) de novas mortes relacionadas a COVID-19
* *total_cases_per_million* : total de casos de COVID-19 por milhão de pessoas
* *new_cases_per_million* : total de novos casos de COVID-19 por milhão de pessoas
* *new_cases_smoothed_per_million* : média móvel (7 dias) de novos casos por milhão de pessoas
* *total_deaths_per_million* : total de mortes relacionadas ao COVID-19 por milhão de pessoas
* *new_deaths_per_million* : total de novas mortes relacionadas ao COVID-19 por milhão de pessoas
* *new_deaths_smoothed_per_million* : médial móvel (7 dias) do número de mortes relacionadas a COVID-19 por milhão de pessoas
* *reproduction_rate* : estimativa da taxa de reprodução do vírus da COVID-19, ou seja, a quantidade de pessoas que um infectado pode contagiar
* *icu_patients* : número de pacientes com COVID-19 na UTI no dia registrado
* *icu_patients_per_million* : número de pacientes com COVID-19 na UTI no dia registrado por milhão de pessoas
* *hosp_patients* : número de pacientes com COVID-19 no dia registrado
* *hosp_patients_per_million* : número de pacientes com COVID-19 no dia registrado por milhão de pessoas
* *weekly_icu_admissions* : número de novos pacientes com COVID-19 que ingressaram na UTI na semana registrada
* *weekly_icu_admissions_per_million* : número de novos pacientes com COVID-19 que ingressaram na UTI na semana registrada por milhão de pessoas
* *weekly_hosp_admissions* : número de novos pacientes com COVID-19 que ingressaram em hospitais na semana registrada
* *weekly_hosp_admissions_per_million* : número de novos pacientes com COVID-19 que ingressaram em hospitais na semana registrada por milhão de pessoas
* *new_tests* : novos testes de COVID-19 realizados(apenas calculados em dias consecutivos)
* *total_tests* : total de testes realizado de COVID-19
* *total_tests_per_thousand* : total de testes realizados de COVID-19 por mil pessoas
* *new_tests_per_thousand* : total de novos testes realizados de COVID-19 por mil pessoas
* *new_tests_smoothed* : média móvel (7 dias) de novos testes registrados
* *new_tests_smoothed_per_thousand* : média móvel (7 dias) de novos testes registrados por milhares de pessoas
* *positive_rate* : taxa de COVID-19 que deram resultados positivos para a doença
* *tests_per_case* : quantidade de testes de COVID-19 realizados para cada caso positivo
* *tests_units* : unidade de medida usada pelo país para se referenciar aos testes
* *total_vaccinations* : total de doses de vacinas administradas
* *people_vaccinated* : quantidade de pessoas que receberam ao menos uma dose da vacina
* *people_fully_vaccinated* : quantidade de pessoas que receberam todas as doses que foram prescritas nos protocolas das vacinas
* *total_boosters* : número de doses de reforço da vacina, ou seja, doses que vão além das prescritas pelos protocolos
* *new_vaccinations* : novas doses de vacinas aplicadas (somente considerando os dias consecutivos)
* *new_vaccinations_smoothed* : média móvel (7 dias) da aplicação de novas doses de vacina
* *total_vaccinations_per_hundred* : total de doses de vacinas administradas a cada 100 pessoas
* *people_vaccinated_per_hundred* : quantidade de pessoas que receberam ao menos uma dose da vacina a cade 100 pessoas
* *people_fully_vaccinated_per_hundred* : quantidades de pessoas que receberam todas as doses de acordo com o protocolo das vacinas a cada 100 pessoas
* *total_boosters_per_hundred* : total de doses de reforço (doses que vão além das protocolares) a cada 100 pessoas
* *new_vaccinations_smoothed_per_million* : média móvel (7 dias) de novas doses administradas a cada 1 milhão de pessoas
* *stringency_index* : Índice que avalia a capacidade de resposta do governo para tentar conter a pandemia (ex: lockdown, fechamento de bares, etc). Este índice vai de 0 a 100
* *population* : População total medida em 2020
* *population_density* : Densidade demográfica
* *median_age* : Média de idade da população baseado na projeção das Nações Unidas para o ano de 2020
* *aged_65_older* : fração da população que tem 65 anos ou mais
* *aged_70_older* : fração da população que tem 70 anos ou mais
* *gdp_per_capita* : Renda per capita
* *extreme_poverty* : fração da população que vive em condições de extrema pobreaza
* *cardiovasc_death_rate* : taxa de mortes ocasionadas por doenças cardiovasculares a cade 100 mil pessoas
* *diabetes_prevalence* : Porcentagem da população (entre 20 e 79 anos) diagnosticadas com diabetes em 2017
* *female_smokers* : fração de mulheres que fumam
* *male_smokers* : fração de homens que fumam
* *handwashing_facilities* : fração da população que tem acesso a condições mínimas de higiene
* *hospital_beds_per_thousand* : leitos de hospital disponível a cada mil pessoas
* *life_expectancy* : Expectativa de vida
* *human_development_index* : Índice de Desenvolvimento Humano (IDH)
* *excess_mortality* : excesso de mortalidade, ou seja, é a diferença entre o número de mortes observados em um tempo específico e o número de mortes que eram esperadas no mesmo período. [Definição de excess mortality](https://www.cdc.gov/nchs/nvss/vsrr/covid19/excess_deaths.html)
* *date* : data de registro da informação

*Como o dataset abrange muitas datas, inicialmente vamos analisar um dia específico (31/08/21) para verificar a relação entre o número de infectados (testes positivos) e número de mortes relacionadas ao COVID-19.*

*Este dia tem boa representatividade, já que está em uma data xx anos após a primeira detecção da doença na China.*

```
#definindo uma função para colocar os rótulos de dados em cima de cada barra
def addlabels(x,y,eixo):
  for i in range(len(x)):
    if eixo == 0:
      #se for do primeiro gráfico temos de adicionar um valor maior para o Y, porque existem mais casos do que mortes
      ax[eixo].text(i,y[i] + 1000000,'{:,}'.format(int(y[i])).replace(',','.'),ha='center',fontsize=14)
    else:
      #se for do segundo gráfico temos de adicionar um valor menor para o Y, porque existem menos mortes do que casos de COVID-19
      ax[eixo].text(i,y[i] + 20000,'{:,}'.format(int(y[i])).replace(',','.'),ha='center',fontsize=14)
```

```
#Avaliando os 10 paises tem mais casos no dia 31/08/2021
dia = '31/08/2021'

#criando o figure e o ax no matplotlib
fig,ax = plt.subplots(figsize=(20,8),nrows=2,ncols=1)

#preparando os dados dos países com mais casos
df_paises_com_mais_casos = df_lista_sem_continentes.loc['2021-08-31'][['location','total_cases']].sort_values(
    by='total_cases',ascending=False)[:10]

#preparando os dados dos países com mais mortes
df_paises_com_mais_mortes = df_lista_sem_continentes.loc['2021-08-31'][['location','total_deaths']].sort_values(
    by='total_deaths',ascending=False)[:10]


#ajustando os parâmetros para o primeiro gráfico
ax[0].bar(data=df_paises_com_mais_casos,x='location',height='total_cases',color='grey')
ax[0].set_title('Casos acumulados de COVID-19 até ' + dia,loc='left',pad=25,fontdict={'fontsize':18})
ax[0].bar(data=df_paises_com_mais_casos[df_paises_com_mais_casos['location'] == 'Brazil'],x='location',height='total_cases',color=(0.2, 0.4, 0.6, 0.6))
ax[0].tick_params(axis='x',labelsize=13,pad=10)

#colocando o rótulo de dados para o primeiro gráfico
addlabels(df_paises_com_mais_casos['location'],df_paises_com_mais_casos['total_cases'],0)

#retirando o eixo y e background de fundo
ax[0].set_frame_on(False)
ax[0].get_yaxis().set_visible(False)

#ajustando os parâmetros para o segundo gráfico
ax[1].bar(data=df_paises_com_mais_mortes,x='location',height='total_deaths',color='grey')
ax[1].set_title('Casos de mortes relacionadas a COVID-19 acumuladas até '+ dia,loc='left',pad=25,fontdict={'fontsize':18,'Family':'Dejavu Sans'})
ax[1].bar(data=df_paises_com_mais_mortes[df_paises_com_mais_mortes['location'] == 'Brazil'],x='location',height='total_deaths',color=(0.2, 0.4, 0.6, 0.6))
ax[1].tick_params(axis='x',labelsize=13,pad=10)

#colocando o rótulo de dados para o segundo gráfico
addlabels(df_paises_com_mais_mortes['location'],df_paises_com_mais_mortes['total_deaths'],1)

#retirando o eixo y e background de fundo
ax[1].get_yaxis().set_visible(False)
ax[1].set_frame_on(False)

#plotando os gráficos
plt.tight_layout(pad=4.0);
```

*Observando os gráficos acima, percebemos que os países que têm mais casos são também os que tem mais mortes. Um fato interessante a notar é a letalidade deste vírus no México, já que embora este país não se encontrava entre os 10 que mais possuiam pessoas com esta doença, ele estava em quarto lugar no ranking de mortes.*

```
#criando o figure e o ax no matplotlib
fig,ax = plt.subplots(figsize=(22,8),nrows=1,ncols=2)

#separando os valores para os três países com mais mortes
brazil = df_lista_sem_continentes[df_lista_sem_continentes['location'] == 'Brazil'].loc['2021-08-31']
usa = df_lista_sem_continentes[df_lista_sem_continentes['location'] == 'United States'].loc['2021-08-31']
india = df_lista_sem_continentes[df_lista_sem_continentes['location'] == 'India'].loc['2021-08-31']


#preparando o primeiro gráfico
df_lista_sem_continentes.loc['2021-08-31'].plot.scatter(x='gdp_per_capita',y='total_deaths',ax=ax[0])
ax[0].set_xlabel('Renda per Capita',labelpad=12)
ax[0].set_ylabel(None)
ax[0].set_title('Renda per Capita e Total de Mortos - 31/08/21',loc='left',fontdict={'fontsize':18},pad=20)

#colocando a label para o Brazil no primeiro gráfico
x_brazil,y_brazil = brazil['gdp_per_capita'],brazil['total_deaths']
ax[0].annotate(xy=(x_brazil,y_brazil),xytext=(x_brazil+6000,y_brazil+8000), arrowprops=dict(arrowstyle="simple",
                            fc="0.6", ec="none",
                            connectionstyle="arc3,rad=0.3"),s='Brazil')

#colocando a label para os Estados unidos no primeiro gráfico
x_usa,y_usa = usa['gdp_per_capita'],usa['total_deaths']
ax[0].annotate(xy=(x_usa,y_usa),xytext=(x_usa+6000,y_usa+8000), arrowprops=dict(arrowstyle="simple",
                            fc="0.6", ec="none",
                            connectionstyle="arc3,rad=0.3"),s='United States')

#colocando a label para a India no primeiro gráfico
x_india,y_india = india['gdp_per_capita'],india['total_deaths']
ax[0].annotate(xy=(x_india,y_india),xytext=(x_india+6000,y_india+8000), arrowprops=dict(arrowstyle="simple",
                            fc="0.6", ec="none",
                            connectionstyle="arc3,rad=0.3"),s='India')

#ajustando o grid do primeiro gráfico
ax[0].grid(linestyle='--',lw=0.5,aa=True)


#preparando o segundo gráfico
df_lista_sem_continentes.loc['2021-08-31'].plot.scatter(x='human_development_index',y='total_deaths',ax=ax[1])
ax[1].set_xlabel('Índice de Desenvolvimento Humano',labelpad=12,fontdict={'fontsize':12})
ax[1].set_ylabel(None)
ax[1].set_title('IDH e Total de Mortos - 31/08/21',loc='left',fontdict={'fontsize':18},pad=20)

#colocando a label para o Brazil no segundo gráfico
x_brazil,y_brazil = brazil['human_development_index'],brazil['total_deaths']
ax[1].annotate(xy=(x_brazil,y_brazil),xytext=(x_brazil+0.03,y_brazil+10000), arrowprops=dict(arrowstyle="simple",
                            fc="0.6", ec="none",
                            connectionstyle="arc3,rad=0.3"),s='Brazil')

#colocando a label para os Estados unidos no segundo gráfico
x_usa,y_usa = usa['human_development_index'],usa['total_deaths']
ax[1].annotate(xy=(x_usa,y_usa),xytext=(x_usa+0.03,y_usa+200), arrowprops=dict(arrowstyle="simple",
                            fc="0.6", ec="none",
                            connectionstyle="arc3,rad=0.3"),s='United \nStates')

#colocando a label para a India no primeiro gráfico
x_india,y_india = india['human_development_index'],india['total_deaths']
ax[1].annotate(xy=(x_india,y_india),xytext=(x_india+0.03,y_india+800), arrowprops=dict(arrowstyle="simple",
                            fc="0.6", ec="none",
                            connectionstyle="arc3,rad=0.3"),s='India')

#ajustando o grid do segundo gráfico
ax[1].grid(linestyle='-.',lw=0.5,aa=True)

#plotando os gráficos

plt.tight_layout(pad=2.0);
```

*Estes gráficos de dispersão nos demonstram que esta doença atingiu países pobres e ricos, não se restringindo apenas à aqueles que não tem as mínimas condições de higiene. Isso se deve ao fato do vírus se espalhar pelas vias aéreas e ser de difícil combate.*
